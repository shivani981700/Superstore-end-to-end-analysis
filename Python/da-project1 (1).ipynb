```python
import pandas as pd

df = pd.read_csv("Sample - Superstore.csv", encoding="latin-1")

# --- cleaning steps ---
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["ship_date"]  = pd.to_datetime(df["ship_date"], errors="coerce")

for c in ["sales", "profit", "discount"]:
    df[c] = pd.to_numeric(df[c], errors="coerce")

df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce").astype("Int64")

# drop rows missing key fields
df = df.dropna(subset=["order_id","order_date","customer_id","product_id"])

# ship date check
df = df[df["ship_date"] >= df["order_date"]]

df_clean = df.copy()   # this is your final cleaned dataframe
```

```python
from sqlalchemy import create_engine

user = "root"
password= "0129"
host = "localhost"
port = 3306
db = "superstore"

engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{db}")
```

```python
df_clean.to_sql(
    name="stg_superstore_clean",
    con=engine,
    if_exists="replace",   # overwrites table each run (best for Day 1)
    index=False,
    chunksize=10000
)
print("Loaded:", len(df_clean), "rows")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='discount', y='profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

# My analysis shows that orders with discounts above 30% frequently generate negative profit, indicating that aggressive discounting is eroding margins rather than driving sustainable growth. A controlled discount policy or targeted discounts for high‑value customers could help improve overall profit margins.

```python
cat = df.groupby('category')[['sales','profit']].sum().reset_index()

cat.plot(x='category', kind='bar', figsize=(8,5))
plt.title("Revenue & Profit by Category")
plt.show()
```

 # Technology is the top revenue‑generating category and also delivers the highest profit, while Furniture shows comparatively low profit despite strong sales, indicating weaker margins. Focusing on margin improvement in Furniture and expanding high‑performing Technology products could increase overall profitability.

```python
region = df.groupby('region')[['sales','profit']].sum().reset_index()

region.plot(x='region', kind='bar', figsize=(8,5))
plt.title("Region Revenue & Profit")
plt.show()
```

 # The West region contributes the highest profit and revenue, whereas the Central region shows weaker profitability despite moderate sales. Targeted marketing and pricing adjustments in lower‑margin regions could help balance overall performance.

```python
orders_per_customer = df.groupby('customer_id')['order_id'].nunique()
repeat = (orders_per_customer > 1).sum()
new = (orders_per_customer == 1).sum()

print("Repeat Customers:", repeat)
print("New Customers:", new)
```

 # Approximately 98% of customers are repeat buyers, indicating extremely strong customer retention. While retention is excellent, increasing new customer acquisition could support long‑term growth and reduce dependency on repeat buyers.
